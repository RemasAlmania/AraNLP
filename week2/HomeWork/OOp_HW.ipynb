```python
import gensim.downloader
from typing import List
from sklearn.metrics.pairwise import cosine_similarity
from numpy import argmax
```

```python

class IRSystem:

    def __init__(self, model_name: str = 'glove-twitter-25'):
        """Load the embedding model."""
        self.model = gensim.downloader.load(model_name)
        self.database = []
        self.database_embeddings = []

    def embed(self, text: str) -> List[float]:

        words = text.split()
        vectors = []
        for word in words:
            if word in self.model:
                vectors.append(self.model[word])
        if not vectors:
            return [0.0] * self.model.vector_size
        final = sum(vectors) / len(vectors)
        return final.tolist()

    def add_documents(self, documents: List[str]) -> None:


        self.database = documents
        self.database_embeddings = [self.embed(doc) for doc in documents]

    def search(self, query: str) -> str:

        query_embedding = self.embed(query)
        scores = cosine_similarity([query_embedding], self.database_embeddings)
        best_index = argmax(scores).item()
        return self.database[best_index]



ir = IRSystem()

ir.add_documents([
    "serry was here yesterday",
    "obama did not die",
    "mark is a lizard",
    "girls no answer",
    "twitter is bad"
])

query = "when did obama die"
result = ir.search(query)
print(f"Query: {query}")
print(f"Result: {result}")
```
