The `else` clause provides an alternative block of code that executes when the `if` condition is `False`. It must follow an `if` statement and handles the case when the condition is not met. Together, `if` and `else` allow two-way branching.

Chained conditionals use `if`, `elif` (else-if), and `else` to handle multiple cases. `elif` allows you to check additional conditions after the initial `if`. Only one branch executes - the first condition that is `True`. This enables multi-way branching.


````python
# Easy question (5.6): Complete the chained conditional.

x = 0
if x > 0:
    print("positive")
elif x < 0:  # use elif
    print("negative")
else:
    print("zero")
````

A nested conditional has an `if` statement inside another `if` block. This allows you to check multiple conditions in sequence. The inner `if` only executes if the outer condition is `True`. Nested conditionals enable complex decision-making logic.

Recursion is when a function calls itself to solve a smaller version of the same problem. A recursive function must have a base case (stopping condition) and a recursive case (calls itself with modified input). Recursion is powerful for problems that have a natural recursive structure.


````python
# Easy question (5.8): What is the key requirement for a recursive function to stop?
# Answer in a comment.

# A recursive function must have a base case.
````
