### 📝 The Exercise: TF-IDF from Scratch

**Instructions for Students:** Complete the `# TODO` sections in the script below. You will need the `math` library to calculate the logarithm for the IDF component.

```python
import math
import string

def tokenize(text):
    """Helper to clean and split text."""
    clean = text.translate(str.maketrans('', '', string.punctuation)).lower()
    return clean.split()

def compute_tf_idf(corpus):
    # 1. Pre-process all documents
    tokenized_docs = [tokenize(doc) for doc in corpus]

    # 2. Build Vocabulary
    all_words = [word for doc in tokenized_docs for word in doc]
    vocab = sorted(list(set(all_words)))
    num_docs = len(corpus)

    # --- STEP A: Compute Term Frequency (TF) ---
    # TF = (Frequency of word in doc) / (Total words in doc)
    all_tf = []
    for doc in tokenized_docs:
        doc_tf = {}
        # TODO: Calculate the count of each word in the current doc
        # TODO: Divide that count by len(doc) to get the TF weight
        # doc_tf[word] = count / total_words
        total_words = len(doc)
        for word in doc:
            count = doc.count(word)
            doc_tf[word] = count / total_words
        all_tf.append(doc_tf)

    # --- STEP B: Compute Inverse Document Frequency (IDF) ---
    # IDF = log(Total Docs / Docs containing the word)
    idf_dict = {}
    for word in vocab:
        # TODO: Count how many documents contain this specific 'word'
        docs_with_word = 0
        for doc in tokenized_docs:
            if word in doc:
                docs_with_word += 1
        # TODO: Calculate IDF using math.log()
        # idf_dict[word] = math.log(num_docs / (1 + docs_with_word))
        # Note: We usually use (1 + docs_with_word) to avoid division by zero!
        idf_dict[word] = math.log(num_docs / (1 + docs_with_word))

    # --- STEP C: Compute Final TF-IDF Matrix ---
    # TF-IDF = TF * IDF
    tfidf_matrix = []
    for doc_tf in all_tf:
        row = []
        for word in vocab:
            # TODO: Get the TF value for this word in this doc (default to 0)
            # TODO: Multiply it by the pre-calculated IDF value for this word
            # TODO: Append the result to the 'row' list
            tf = doc_tf.get(word, 0)
            idf = idf_dict[word]
            row.append(tf * idf)
        tfidf_matrix.append(row)

    return tfidf_matrix, vocab

# --- TEST DATA ---
my_corpus = [
    "The cat is on the mat",
    "The dog is on the rug",
    "The cat and the dog are friends"
]

matrix, features = compute_tf_idf(my_corpus)

print("TF-IDF Matrix:")
print(features)
for row in matrix:
    print(row)
```

### 💡 The Mathematical Hint for Students

1. **Term Frequency (TF):**
$$TF(t, d) = \\frac{\\text{Count of term } t \\text{ in document } d}{\\text{Total number of terms in document } d}$$

2. **Inverse Document Frequency (IDF):**
$$IDF(t) = \\log\\left(\\frac{\\text{Total number of documents}}{\\text{Number of documents with term } t}\\right)$$

3. **TF-IDF:**
$$TF\\text{-}IDF = TF \\times IDF$$
